```python
#Hence will consider the Model2 as our model to move forward with as each variable seems to significant
#Calculating VIF 
#Lets try to find out row wise outliers to improve model performance 
```

```python
#VIF
rsq_age = smf.ols('age~KM+HP+cc+Doors+Gears+Qtax+Weight', data=toco1).fit().rsquared
vif_age = 1/(1-rsq_age)

rsq_KM = smf.ols('KM~age+HP+cc+Doors+Gears+Qtax+Weight', data=toco1).fit().rsquared
vif_KM = 1/(1-rsq_KM)

rsq_HP = smf.ols('HP~age+KM+cc+Doors+Gears+Qtax+Weight', data=toco1).fit().rsquared
vif_HP = 1/(1-rsq_HP)

rsq_cc = smf.ols('cc~age+KM+HP+Doors+Gears+Qtax+Weight', data=toco1).fit().rsquared
vif_cc = 1/(1-rsq_cc)

rsq_Doors = smf.ols('Doors~age+KM+HP+cc+Gears+Qtax+Weight', data=toco1).fit().rsquared
vif_Doors = 1/(1-rsq_Doors)

rsq_Gears = smf.ols('Gears~age+KM+HP+cc+Doors+Qtax+Weight', data=toco1).fit().rsquared
vif_Gears = 1/(1-rsq_Gears)

rsq_Qtax = smf.ols('Qtax~age+KM+HP+cc+Doors+Gears+Weight', data=toco1).fit().rsquared
vif_Qtax = 1/(1-rsq_Qtax)

rsq_Weight = smf.ols('Weight~age+KM+HP+cc+Doors+Gears+Qtax', data=toco1).fit().rsquared
vif_Weight = 1/(1-rsq_Weight)

d1 = {'variables':['age','KM','HP','cc','Doors','Gears','Qtax','Weight'], 'VIF':[vif_age, vif_KM, vif_HP, vif_cc, vif_Doors, vif_Gears, vif_Qtax, vif_Weight]}
vif_df = pd.DataFrame(d1)
vif_df
```

```python
#Test for Homoscedasticity or Heteroscedasticity (plotting model's standardized fitted values vs standardized residule values)
def get_standardized_values(vals):
    return (vals - vals.mean())/vals.std()
```
